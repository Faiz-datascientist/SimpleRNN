```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=["The cat sat quietly by the window, watching the birds outside.",
"She opened the book and lost herself in its captivating story.",
"The city skyline glowed brightly under the evening sky.",
"He prepared a delicious meal using fresh ingredients from the garden.",
"Learning a new language can be challenging but rewarding"]
```

```python
sent
```

```python
voc_limitation=10000
```

```python
one_hot_repr=[one_hot(words,voc_limitation) for words in sent]
```

```python
one_hot_repr
```

```python
from tensorflow.keras.layers import Embedding
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
from tensorflow.keras.models import Sequential
```

```python
max([len(i) for i in one_hot_repr])
```

```python
max_length=15
```

```python
embedded_docs=pad_sequences(one_hot_repr,maxlen=max_length,padding='post')
```

```python
print(embedded_docs)    
```

```python
#feature representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_limitation,dim,input_length=max_length))
model.compile(optimizer='adam',loss='mean_squared_error',metrics='mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
